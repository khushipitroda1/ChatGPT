```python
import gradio as gr
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition")
```

```python
def transcribe1(audio_file):
    text= p(audio_file)["text"]
    response = openai.Completion.create(model="text-davinci-003", prompt=text, temperature=0, max_tokens=7)
    audio_array = generate_audio(response["choices"][0].text.replace('\n', ''), history_prompt="v2/en_speaker_9")
    write_wav("bark_generation.wav", SAMPLE_RATE, audio_array)
    return [text,"bark_generation.wav"]
```

```python
output_1= gr.Textbox(label="Speech to text")
output_2= gr.Audio("bark_generation.wav")
```

```python
gr.Interface(
    fn=transcribe1, 
    inputs=gr.Audio(source="microphone", type="filepath"), 
    outputs=[output_1,output_2],live=True).launch()
```
